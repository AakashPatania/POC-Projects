```python
import psycopg2 
import pandas as pd
```

```python
def create_database():
    # Connecting to default database
    conn = psycopg2.connect("host=localhost dbname=postgres user=postgres password=root port=5433")
    conn.set_session(autocommit=True)
    cur = conn.cursor()
    
    # Creating a database with utf8 encoding
    
    cur.execute("DROP DATABASE accounts")
    cur.execute("CREATE DATABASE accounts")
    
    conn.close()
    
    # Connecting to sparkify database
    
    conn = psycopg2.connect("host=localhost dbname=accounts user=postgres password=root port=5433")
    
    cur = conn.cursor()
    
    return cur, conn
```
